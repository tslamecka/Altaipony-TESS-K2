```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib import rcParams

rcParams['font.family'] = 'serif'
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':10})

plt.rcParams['xtick.major.size'] = 8  # Length of major x-ticks
plt.rcParams['xtick.minor.size'] = 4   # Length of minor x-ticks
plt.rcParams['ytick.major.size'] = 8  # Length of major y-ticks
plt.rcParams['ytick.minor.size'] = 4   # Length of minor y-ticks
```

```python
dfk = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\K2_catalogues\K2_catalogue_cut_revised.csv")
dft = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\TESS_catalogues\TESS_catalogue_cut_revised.csv")
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(dfk['Average percentage increase'], dfk['Flare #'], c = "mediumpurple", edgecolor = "black", s = 25, label = "K2")
plt.scatter(dft['Average percentage increase'], dft['Flare #'], c = "limegreen", edgecolor = "black", s = 25, label = "TESS")
plt.xscale('log')
plt.xlabel("Average Percentage Increase")
plt.ylabel("Number of Flares")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(dfk['Teff'], dfk['Flare #'], c = "mediumpurple", edgecolor = "black", s = 25, label = "K2")
plt.scatter(dft['Teff'], dft['Flare #'], c = "limegreen", edgecolor = "black", s = 25, label = "TESS")
plt.xlabel("Effective Temperature (K)")
plt.ylabel("Number of Flares")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(dfk['logg'], dfk['Flare #'], c = "mediumpurple", edgecolor = "black", s = 25, label = "K2")
plt.scatter(dft['logg'], dft['Flare #'], c = "limegreen", edgecolor = "black", s = 25, label = "TESS")
plt.xlabel("log g")
plt.ylabel("Number of Flares")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(dfk['Teff'], dfk['Average percentage increase'], c = "mediumpurple", edgecolor = "black", s = 25, label = "K2")
plt.scatter(dft['Teff'], dft['Average percentage increase'], c = "limegreen", edgecolor = "black", s = 25, label = "TESS")
plt.yscale('log')
plt.xlabel("Effective Temperature (K)")
plt.ylabel("Average Percentage Increase")
plt.legend()
plt.show()
```
