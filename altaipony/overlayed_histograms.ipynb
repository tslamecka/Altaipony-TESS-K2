```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib import rcParams

rcParams['font.family'] = 'serif'
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':10})

plt.rcParams['xtick.major.size'] = 8  # Length of major x-ticks
plt.rcParams['xtick.minor.size'] = 4   # Length of minor x-ticks
plt.rcParams['ytick.major.size'] = 8  # Length of major y-ticks
plt.rcParams['ytick.minor.size'] = 4   # Length of minor y-ticks
```

```python
dfk = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\K2_catalogues\K2_catalogue_cut_revised.csv")
dft = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\TESS_catalogues\TESS_catalogue_cut_revised.csv")
```

```python
plt.figure(figsize=(10, 5), dpi = 200)

plt.hist(dfk['Flares per day'], width = 0.03, bins = 30, color = "mediumpurple", edgecolor = "black", label = "K2")
plt.hist(dft['Flares per day'], width = 0.03, bins = 30, color = "limegreen", edgecolor = "black", label = "TESS")

plt.xlabel("Flares per day")
plt.ylabel("Frequency")

#plt.yscale('log')

plt.xlim(-0.075,1.6)

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)

plt.hist(dfk['Average percentage increase'], bins = 500, color = "mediumpurple", edgecolor = "black", label = "K2")
plt.hist(dft['Average percentage increase'], bins = 500, color = "limegreen", edgecolor = "black", label = "TESS")

plt.xlabel("Average percentage increase")
plt.ylabel("Frequency")

plt.yscale('log')
plt.xscale('log')

#plt.xlim(-0.075,1.6)

plt.legend()

plt.show()
```
