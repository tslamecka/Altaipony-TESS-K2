```python
def get_epic_id(ra, dec):
    coords = SkyCoord(ra = ra, dec = dec, unit = (u.deg, u.deg))
    r = 5 * u.arcsec
    
    obs = Observations.query_region(coords, radius = r)
    
    if len(obs) > 0:
        return obs[0]['target_name']
    return None

dft_cut['EPIC_ID'] = dft_cut.apply(lambda row: get_epic_id(row['RA'], row['Dec']), axis=1)
```
