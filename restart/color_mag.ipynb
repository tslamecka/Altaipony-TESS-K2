```python
import pandas as pd
pd.set_option("display.max_columns", None)
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib import rcParams

rcParams['font.family'] = 'serif'
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':10})

plt.rcParams['xtick.major.size'] = 8  # Length of major x-ticks
plt.rcParams['xtick.minor.size'] = 4   # Length of minor x-ticks
plt.rcParams['ytick.major.size'] = 8  # Length of major y-ticks
plt.rcParams['ytick.minor.size'] = 4   # Length of minor y-ticks
```

```python
gaia = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\\restart\gaia_epoch2016.csv")
```

```python
gaia.head()
```

Columns to keep: EPIC_ID, TIC_ID, RA, DEC, Plx (mac), Gmag, AG, BP-RP, Teff, logg, E(BP-RP)

```python
cut = pd.DataFrame(columns = ['EPIC_ID', 'TIC_ID', 'RA', 'DEC', 'Plx', 'Gmag', 'AG', 'BP-RP', 'E(BP-RP)', 'Teff', 'logg'])
```

```python
cut['EPIC_ID'] = gaia['EPIC_ID']
cut['TIC_ID'] = gaia['TIC_ID']
cut['RA'] = gaia['RA']
cut['DEC'] = gaia['DEC']
cut['Plx'] = gaia['Plx']
cut['Gmag'] = gaia['Gmag']
cut['AG'] = gaia['AG']
cut['BP-RP'] = gaia['BP-RP']
cut['E(BP-RP)'] = gaia['E(BP-RP)']
cut['Teff'] = gaia['Teff']
cut['logg'] = gaia['logg']
```

```python
cut.head()
```

```python
print(f'In the original data set, there are ' + str(len(np.where(gaia['Plx'].isnull() == True)[0])) + ' rows with no PARALLAX data')
print('After cutting the data set, there are')
print(str(len(np.where(cut['Gmag'].isnull() == True)[0])) + ' rows without GMAG data')
print(str(len(np.where(cut['AG'].isnull() == True)[0])) + ' rows without AG data')
print(str(len(np.where(cut['BP-RP'].isnull() == True)[0])) + ' rows with no BP-RP data')
print(str(len(np.where(cut['E(BP-RP)'].isnull() == True)[0])) + ' rows without E(RP-BP) data')
```

```python
cut = cut[cut['Plx'].isnull() == False]
df = cut[cut['Gmag'].isnull() == False]
final = df[df['BP-RP'].isnull() == False]
len(final)
```

After GAIA crossmatch: 47730 \
After removing Plx nulls (7272): 40458 \
After removing Gmag nulls (4): 40454 \
After removing BP-RP nulls (72): 40386

```python
final['AG'] = final['AG'].fillna(0)
final['E(BP-RP)'] = final['E(BP-RP)'].fillna(0)
```

```python
final['(BP-RP)_0'] = final['BP-RP'] - final['E(BP-RP)']
```

```python
final['Dist'] = 1 / (final['Plx'] / 1000)
```

```python
final['Amag'] = final['Gmag'] - final['AG'] - (5*np.log10(final['Dist']) - 5)
```

```python
final.head()
```

```python
len(final)
```

```python
plt.figure(figsize=(10, 5), dpi = 200)

plt.scatter(final['(BP-RP)_0'], final['Amag'], c = "black", s = 0.1)

plt.xlabel("$(BP-RP)_0$")
plt.ylabel("Absolute Magnitude")

#plt.xlim(-1,4)

plt.gca().invert_yaxis()

plt.tick_params(direction='in', which='both')
plt.minorticks_on()

plt.axvline(x = 0.8, color = "black", linestyle = ":", linewidth = 1)
plt.axhline(y = 6, color = "black", linestyle = ":", linewidth = 1)

plt.title("Color Magnitude Diagram for Stars Observed in TESS and Kepler")

plt.show()
```

```python
mdwarf_cut = final[final['(BP-RP)_0'] > 0.8]
mdwarf_cut = mdwarf_cut[mdwarf_cut['Amag'] > 6]
```
