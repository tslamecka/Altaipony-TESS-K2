```python
from altaipony.lcio import from_mast
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.5f' % x)
%matplotlib inline
```

```python
dfk = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\K2_catalogue_degrees.csv")
dft = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\TESS_catalogue_degrees_cut.csv")
```

```python
dfk.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
dft = dft.drop(dft.columns[-1], axis=1)
```

```python
columns = ['Unnamed: 0', 'TIC ID', 'Teff', 'logg', 'Flare #', 'Observation time', 'Average flare durartion', 'Longest flare duration', 'Shortest flare duration', 'Average percentage increase', 'Highest percentage increase', 'Lowest percentage increase', 'Flares per hour', 'Flares per day', 'RA', 'Dec']
empty = pd.DataFrame(columns = columns, index = range(2150, 3242))
dft = pd.concat([dft, empty])
```

```python
def remove_last_decimal(x):
    if pd.notnull(x):
        return int(x * 10**5) / 10**5
    else:
        return x
```

```python
dft['Dec'] = dft['Dec'].apply(remove_last_decimal)
```

```python
dft.to_csv("C:\Vs_code_projects\mdwarf_flares\TESS_catalogue_degrees_cut.csv", index = False)
```
