```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib import rcParams

rcParams['font.family'] = 'serif'
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':10})

plt.rcParams['xtick.major.size'] = 8  # Length of major x-ticks
plt.rcParams['xtick.minor.size'] = 4   # Length of minor x-ticks
plt.rcParams['ytick.major.size'] = 8  # Length of major y-ticks
plt.rcParams['ytick.minor.size'] = 4   # Length of minor y-ticks
```

```python
df = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\TESS_catalogue_revised.csv")
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(df['Average percentage increase'], df['Flare #'], c = "limegreen", edgecolor = "black", s = 25, label = "TESS")
plt.xscale('log')
plt.xlabel("Average Percentage Increase")
plt.ylabel("Number of Flares")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(df['Teff'], df['Flare #'], c = "limegreen", edgecolor = "black", s = 25)
plt.xlabel("Effective Temperature (K)")
plt.ylabel("Number of Flares")
#plt.ticklabel_format(style = "sci", axis = "y", scilimits = (0,0)) 
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(df['logg'], df['Flare #'], c = "limegreen", edgecolor = "black", s = 25)
plt.xlabel("log g")
plt.ylabel("Number of Flares")
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(df['Teff'], df['Average percentage increase'], c = "limegreen", edgecolor = "black", s = 25)
plt.xlabel("Effective Temperature (K)")
plt.ylabel("Average Percentage Increase")
plt.yscale("log")
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.scatter(df['logg'], df['Average percentage increase'], c = "limegreen", edgecolor = "black", s = 25)
plt.xlabel("log g")
plt.ylabel("Average Percentage Increase")
plt.yscale("log")
plt.show()
```
