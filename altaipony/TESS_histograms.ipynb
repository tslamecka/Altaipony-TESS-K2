```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from matplotlib import rcParams

rcParams['font.family'] = 'serif'
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':10})

plt.rcParams['xtick.major.size'] = 8  # Length of major x-ticks
plt.rcParams['xtick.minor.size'] = 4   # Length of minor x-ticks
plt.rcParams['ytick.major.size'] = 8  # Length of major y-ticks
plt.rcParams['ytick.minor.size'] = 4   # Length of minor y-ticks
```

```python
df = pd.read_csv("C:\Vs_code_projects\mdwarf_flares\TESS_catalogue_frequency.csv")
```

```python
len(df)
```

```python
df_no_0 = df[df["Flare #"] > 0]
```

```python
len(df_no_0) - len(df)
```

```python
df_revised = df[df["Lowest percentage increase"] >= 0]
```

```python
df_cut = df_revised[df_revised["Highest percentage increase"] < 100]
```

```python
plt.figure(figsize=(10, 6), dpi = 200)
plt.hist(df['Flare #'], bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Number of Flares")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.hist(df['Teff'], bins = 50, color = "limegreen", edgecolor = "black", label = "TESS")
plt.yscale('log')
plt.xlabel("Effective Temperature")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.hist(df['logg'], bins = 50, color = "limegreen", edgecolor = "black", label = "TESS")
plt.yscale('log')
plt.xlabel("logg")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.hist(df['Observation time'], bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Observation time (days)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 8), dpi = 200)

plt.subplot(3, 1, 1)
plt.hist(df['Average flare duration']*24, bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Average Flare Duration (hours)")
plt.ylabel("Frequency")

plt.subplot(3, 1, 2)
plt.hist(df['Shortest flare duration']*24, bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Shortest Flare Duration (hours)")
plt.ylabel("Frequency")

plt.subplot(3, 1, 3)
plt.hist(df['Longest flare duration']*24, bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Longest Flare Duration (hours)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8), dpi = 200)

plt.subplot(3, 1, 1)
plt.hist(df['Average percentage increase'], bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Average Percentage Increase")
plt.ylabel("Frequency")

plt.subplot(3, 1, 2)
plt.hist(df['Lowest percentage increase'], bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Lowest Percentage Increase")
plt.ylabel("Frequency")

plt.subplot(3, 1, 3)
plt.hist(df['Highest percentage increase'], bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Highest Percentage Pncrease")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8), dpi = 200)

plt.subplot(3, 1, 1)
plt.hist(df_cut['Average percentage increase'], bins = 100, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Average Percentage Increase")
plt.ylabel("Frequency")

plt.subplot(3, 1, 2)
plt.hist(df_cut['Lowest percentage increase'], bins = 100, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Lowest Percentage Increase")
plt.ylabel("Frequency")

plt.subplot(3, 1, 3)
plt.hist(df_cut['Highest percentage increase'], bins = 100, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Highest Percentage Pncrease")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.hist(df['Flares per hour'], bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Flares per hour - TESS")
plt.ylabel("Frequency")
plt.xlim(-0.0025,0.07)
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi = 200)
plt.hist(df['Flares per day'], bins = 50, color = "limegreen", edgecolor = "black")
plt.yscale('log')
plt.xlabel("Flares per day - TESS")
plt.ylabel("Frequency")
plt.xlim(-0.075,1.6)
plt.show()
```
